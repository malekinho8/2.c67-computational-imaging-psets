```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lb

plt.rc('text',usetex=True)
plt.rc('font',family='serif')
plt.rc('font',size=10)
```

```python
def f(x,L_1=11,L_2=10):
    return 1/2 * (np.cos(2*np.pi*1/L_1*x) + np.cos(2*np.pi*1/L_2*x))
```

```python
x = np.linspace(-50,50,10000)
y = f(x,11,2)
```

```python
plt.plot(x,y)
```

```python
from numpy.fft import ifft2,fft2,fft,ifft,fftfreq, fftshift, ifftshift
```

```python
F_nu = fftshift(fft(y))
freqs = F_nu.shape[0] * fftshift(fftfreq(F_nu.shape[0]))

print(F_nu.shape)
```

```python
plt.plot(freqs,np.abs(F_nu))
plt.xlim(-100,100)
plt.xlabel('Wavenumber $k_x$')
```
