## PS1.1 - Linear 1D Forward Operator

```python
# import the necessary packages
import numpy as np
import matplotlib.pyplot as plt

# use latex font and serif font for plots
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# set figure background color to white
plt.rcParams['figure.facecolor'] = 'white'
```

### Variable Definitions

```python
# first enter the information for the H matrix
H = np.array([
    [0.5, 0.3214, 0.0667, 0.0313, 0, 0, 0, 0, 0],
    [0.5, 0.3571, 0.2667, 0.0938, 0.0588, 0, 0, 0, 0],
    [0, 0.3214, 0.3333, 0.2188, 0.1176, 0, 0, 0, 0],
    [0, 0, 0.2667, 0.3125, 0.1765, 0.0333, 0, 0, 0],
    [0, 0, 0.0667, 0.2188, 0.2941, 0.3, 0.1562, 0, 0],
    [0, 0, 0, 0.0938, 0.1765, 0.3333, 0.1875, 0, 0],
    [0, 0, 0, 0, 0.0313, 0.1176, 0.3000, 0.3077, 0],
    [0, 0, 0, 0, 0.0588, 0.0333, 0.1875, 0.3846, 0.4737],
    [0, 0, 0, 0, 0, 0, 0.1562, 0.3077, 0.5263]
])

print(H)
```

Looks good!

```python
# consider object who is a collection of 3 light bulbs spaced at locations j=1,5,9
f_1 = np.array([1, 0, 0, 0, 1, 0, 0, 0, 1])
print(f_1)
```

```python
# consider a second object spaced at locations j=1,3,5,7,9
f_2 = np.array([1, 0, 1, 0, 1, 0, 1, 0, 1])
print(f_2)
```

### (a) Calculate the noise-less measurements, $g_n = H f_n$, for $n = 1,~2$, and plot them, as we done in Figure 3.3 Notes.

```python
# compute g_1, i.e. the measured light intensity at each of the 9 locations due to the first object
g_1 = np.matmul(H, f_1)
print(g_1)
```

```python
# compute g_2, i.e. the measured light intensity at each of the 9 locations due to the second object
g_2 = np.matmul(H, f_2)
print(g_2)
```

```python
# plot the two objects and the two measured light intensities, using a 2x2 subplot with bars
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].bar(np.arange(1, 10), f_1)
axs[0, 0].set_title(r'Object 1')
axs[0, 0].set_xlabel(r'Location $j$')
axs[0, 0].set_ylabel(r'Intensity $f_1(j)$')
axs[0, 0].set_xticks(np.arange(1, 10))
axs[0, 0].set_yticks(np.arange(0, 2, 1))

# plot the second object in the second row below the first object
axs[1, 0].bar(np.arange(1, 10), f_2)
axs[1, 0].set_title(r'Object 2')
axs[1, 0].set_xlabel(r'Location $j$')
axs[1, 0].set_ylabel(r'Intensity $f_2(j)$')
axs[1, 0].set_xticks(np.arange(1, 10))
axs[1, 0].set_yticks(np.arange(0, 2, 1))

# plot the first measured light intensity in the first row, second column
axs[0, 1].bar(np.arange(1, 10), g_1)
axs[0, 1].set_title(r'Measured Intensity 1')
axs[0, 1].set_xlabel(r'Location $j$')
axs[0, 1].set_ylabel(r'Intensity $g_1(j)$')
axs[0, 1].set_xticks(np.arange(1, 10))
axs[0, 1].set_yticks(np.arange(0, 2, 1))

# plot the second measured light intensity in the second row, second column
axs[1, 1].bar(np.arange(1, 10), g_2)
axs[1, 1].set_title(r'Measured Intensity 2')
axs[1, 1].set_xlabel(r'Location $j$')
axs[1, 1].set_ylabel(r'Intensity $g_2(j)$')
axs[1, 1].set_xticks(np.arange(1, 10))
axs[1, 1].set_yticks(np.arange(0, 2, 1))

# use a tight layout and show the plot
plt.tight_layout()
plt.show()
```

### (b) Compute the direct linear inverses, $\hat{f}_n = H^{-1} g_n$, for $n = 1,~2$, and plot them.

```python
# calculate H-1
H_inv = np.linalg.inv(H)
print(H_inv)
```

```python
# check that H-1 * H = I
print(np.matmul(H_inv, H))
```

```python
# compute f_hat_1, i.e. the reconstruction of the first object from the first measured light intensity
f_hat_1 = np.matmul(H_inv, g_1)
print(f_hat_1)
```

```python
# compute f_hat_2, i.e. the reconstruction of the second object from the second measured light intensity
f_hat_2 = np.matmul(H_inv, g_2)
print(f_hat_2)
```

```python
# plot the two reconstructed objects and the two measured light intensities, using a 2x2 subplot with bars
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].bar(np.arange(1, 10), f_hat_1)
axs[0, 0].set_title(r'Reconstructed Object 1')
axs[0, 0].set_xlabel(r'Location $j$')
axs[0, 0].set_ylabel(r'Intensity $f_1(j)$')
axs[0, 0].set_xticks(np.arange(1, 10))
axs[0, 0].set_yticks(np.arange(0, 2, 1))

# plot the second object in the second row below the first object
axs[1, 0].bar(np.arange(1, 10), f_hat_2)
axs[1, 0].set_title(r'Reconstructed Object 2')
axs[1, 0].set_xlabel(r'Location $j$')
axs[1, 0].set_ylabel(r'Intensity $f_2(j)$')
axs[1, 0].set_xticks(np.arange(1, 10))
axs[1, 0].set_yticks(np.arange(0, 2, 1))

# plot the first measured light intensity in the first row, second column
axs[0, 1].bar(np.arange(1, 10), g_1)
axs[0, 1].set_title(r'Measured Intensity 1')
axs[0, 1].set_xlabel(r'Location $j$')
axs[0, 1].set_ylabel(r'Intensity $g_1(j)$')
axs[0, 1].set_xticks(np.arange(1, 10))
axs[0, 1].set_yticks(np.arange(0, 2, 1))

# plot the second measured light intensity in the second row, second column
axs[1, 1].bar(np.arange(1, 10), g_2)
axs[1, 1].set_title(r'Measured Intensity 2')
axs[1, 1].set_xlabel(r'Location $j$')
axs[1, 1].set_ylabel(r'Intensity $g_2(j)$')
axs[1, 1].set_xticks(np.arange(1, 10))
axs[1, 1].set_yticks(np.arange(0, 2, 1))

# use a tight layout and show the plot
plt.tight_layout()
plt.show()
```

### (c) Now corrupt the measurements $g_n$ by adding to every element of the vectors a Gaussian random number of mean equal to zero and variance equal to $\sigma^2 = (0.1)^2$. For example, in Matlab this is implemented with the following command: `g noisy = g + 0.1 * randn(9,1)`. Repeat the direct linear inverse calculation with the noisy measurements and plot the results. Comment briefly on what you observe.

```python
# obtain the noisy measured light intensity 1 by adding Gaussian noise to g_1
g_noisy_1 = g_1 + np.random.normal(0, 0.1, 9)
print(g_noisy_1)
```

```python
# obtain the noisy measured light intensity 2 by adding Gaussian noise to g_2
g_noisy_2 = g_2 + np.random.normal(0, 0.1, 9)
print(g_noisy_2)
```

```python
# perform the inverse transform on the noisy measured light intensities
f_hat_noisy_1 = np.matmul(H_inv, g_noisy_1)
f_hat_noisy_2 = np.matmul(H_inv, g_noisy_2)
```

```python
# return the f_hat_noisy latex string using pmatrix (i.e. print(f_hat_noisy_latex) should return the latex string)
f_hat_noisy_1_latex_string = r'\begin{{pmatrix}} {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \end{{pmatrix}}'.format(*f_hat_noisy_1)
f_hat_noisy_2_latex_string = r'\begin{{pmatrix}} {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \\ {} \end{{pmatrix}}'.format(*f_hat_noisy_2)
print(f_hat_noisy_1_latex_string)
print(f_hat_noisy_2_latex_string)
```

```python
# return the noisy latex strings, but with the values rounded to 2 decimal places
f_hat_noisy_1_latex_string_rounded = r'\begin{{pmatrix}} {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \end{{pmatrix}}'.format(*f_hat_noisy_1)
f_hat_noisy_2_latex_string_rounded = r'\begin{{pmatrix}} {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \\ {:.2} \end{{pmatrix}}'.format(*f_hat_noisy_2)
print(f_hat_noisy_1_latex_string_rounded)
print(f_hat_noisy_2_latex_string_rounded)
```
